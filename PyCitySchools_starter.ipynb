```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

school_data

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(2)
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python

total_schools = len(school_data_complete["school_name"].unique())
total_students = school_data_complete["student_name"].count()
total_budget = school_data_complete["budget"].unique().sum()
average_math= school_data_complete["math_score"].mean()
average_reading= school_data_complete["reading_score"].mean()
overall_passing= (average_math + average_reading)/2
Filter_Math_Score= (school_data_complete[school_data_complete["math_score"]>=70])
Sum_Students_after_MathFilter=Filter_Math_Score["student_name"].count()
New_Ave_Math_Pass= (Sum_Students_after_MathFilter/total_students)*100
Filter_Read_Score= (school_data_complete[school_data_complete["reading_score"]>=70])
Sum_Students_after_ReadFilter=Filter_Read_Score["student_name"].count()
New_Ave_Read_Pass= (Sum_Students_after_ReadFilter/total_students)*100
summery= {'Total Schools': [total_schools],'Total Students': [total_students], 'Total Budget':[total_budget], 'Avarage Math Score':[average_math],'Avarage Reading Score':[average_reading],'% Passing Math': [New_Ave_Math_Pass], '% Passing Reading': [New_Ave_Read_Pass],'% overall Passing Score' : [overall_passing] }
district_summary = pd.DataFrame(summery)
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
grouped_School= school_data_complete.groupby(["school_name"])
grouped_School_Students= grouped_School["student_name"].size()
grouped_School_Budget= grouped_School["budget"].first()
Average_Budget_perSchool= grouped_School_Budget/grouped_School_Students
Average_Math_Score= grouped_School["math_score"].mean()
Average_Reading_Score= grouped_School["reading_score"].mean()
Filter_Math_Score= school_data_complete[school_data_complete["math_score"]>=70].groupby(["school_name"]).size()
Math_Score_Percent= (Filter_Math_Score/grouped_School_Students)*100
Filter_Reading_Score= school_data_complete[school_data_complete["reading_score"]>=70].groupby(["school_name"]).size()
Reading_Score_Percent= (Filter_Reading_Score/grouped_School_Students)*100
Overall_Passing_Rate= (Math_Score_Percent + Reading_Score_Percent)/2
school_summery= {'Total Students': grouped_School_Students,'Total School Budget': grouped_School_Budget,
                  'Per Student Budget':Average_Budget_perSchool,
                 'Average Math Score':Average_Math_Score,'Average Reading Score': Average_Reading_Score, 
                 '% Passing Math': Math_Score_Percent,'% Passing Reading' : Reading_Score_Percent, 'Overall Passing Rate': Overall_Passing_Rate }
School_Summary= pd.DataFrame(school_summery)
School_Summary.head(5)
```
